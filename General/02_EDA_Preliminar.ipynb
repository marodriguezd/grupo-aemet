# Fase 01: Exploratory Data Analysis (EDA) Preliminar

Aquí comprobamos la calidad de los datos y generamos algunas visualizaciones estáticas.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuramos un estilo bonito para los gráficos
sns.set_theme(style="whitegrid")
```

```python
# Cargar un archivo de ejemplo (asegúrate de haber ejecutado el ETL y tener datos)
try:
    df = pd.read_csv('climaticos_2024.csv')
    df['fecha'] = pd.to_datetime(df['fecha'])
    print("Muestra de datos cargada:")
    display(df.head())
    
    print("\nTipos del DataFrame:")
    df.info()
except Exception as e:
    print(f"Aún no hay datos para el EDA: {e}")
```

```python
# Visualización de la temperatura media
if 'df' in locals():
    # Filtramos por una estación de ejemplo (Madrid Retiro es 3195)
    df_madrid = df[df['indicativo'] == '3195']
    
    if not df_madrid.empty:
        plt.figure(figsize=(12, 6))
        plt.plot(df_madrid["fecha"], df_madrid["tmed"], label="Temperatura media", color='orange')
        plt.title("Evolución de la temperatura media en Madrid (Retiro) durante 2024")
        plt.xlabel("Fecha")
        plt.ylabel("Temperatura (ºC)")
        plt.legend()
        plt.show()
    else:
        print("No hay datos para la estación 3195 en la muestra.")
```
