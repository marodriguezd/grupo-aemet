```python
# Imports para el proyecto AEMET
import os
from IPython.display import Markdown, display  # prueba Markdown para display
import requests
from dotenv import load_dotenv
import pandas as pd
import numpy as np
import json

# Cargar variables de entorno desde el archivo .env (si existe)
load_dotenv()

# Obtener la clave de la API desde las variables de entorno
API_KEY = os.getenv('AEMET_API_KEY')
if not API_KEY:
    raise RuntimeError('AEMET_API_KEY no encontrada en .env')
```

- Conexion para los datos de las estaciones `/api/valores/climatologicos/inventarioestaciones/todasestaciones`

```python

# Datos Configuración para la consulta a la API de AEMET

api_key = API_KEY  # usar la clave cargada desde .env
url_base = "https://opendata.aemet.es/opendata/api/valores/climatologicos/inventarioestaciones/todasestaciones"
headers = {'cache-control': "no-cache",
           'api_key'      : api_key}

def obtener_inventario_estaciones():

    # 1. Solicitar el enlace de descarga
    response = requests.get(url_base, headers=headers)
    
    if response.status_code == 200:
        meta_datos = response.json()
        print(meta_datos) 
        print(f"Respuesta AEMET: {meta_datos.get('descripcion')}")
        
        if meta_datos.get('estado') == 200:
            url_final = meta_datos.get('datos')
            
            # 2. Descargar el JSON con los datos de las estaciones
            datos_response = requests.get(url_final)
            return datos_response.json()
        else:
            print(f"Error AEMET: {meta_datos.get('descripcion')}")
    else:
        print(f"Error de red: {response.status_code}")
    return None

# Ejecución y visualización
estaciones = obtener_inventario_estaciones()

# Ejemplo: total de estaciones
print(f"Total de estaciones: {len(estaciones)}")   

# Ejemplo: del JSON completo de estaciones
print(estaciones)

if estaciones:
    
    # Ejemplo: Imprimir contenido del JSON
    for est in estaciones:
        
        #Ejemplo de campos de JSON por estacion 
        # {'latitud': '394924N', 'provincia': 'ILLES BALEARS', 'altitud': '490', 'indicativo': 'B013X', 'nombre': 'ESCORCA, LLUC', 'indsinop': '08304', 'longitud': '025309E'}
        print(f"Estación: {est['nombre']} | Provincia: {est['provincia']} | ID: {est['indicativo']}")
```

```python
# Conexion a la API de AEMET para obtener datos climáticos diarios de una estación específica en un rango de fechas determinado.

# Pasando los parámetros a la funcion obtener_valores_climaticos(), para mas adelante poner hacer bucles infomacio por dia.
#      -- fecha_ini_utc.
#      -- fecha_fin_utc .
#      -- idema          .

# Configuración cadena de consulta a la API de AEMET
api_key = API_KEY  # usar la clave desde .env
headers = {'cache-control': "no-cache", 
           'api_key'      : api_key}

def obtener_valores_climaticos(fecha_ini_utc, fecha_fin_utc, idema):
    url_base = f"https://opendata.aemet.es/opendata/api/valores/climatologicos/diarios/datos/fechaini/{fecha_ini_utc}/fechafin/{fecha_fin_utc}/estacion/{idema}"
    
    print ( f"1 ---- paso URL de consulta: {url_base}")
    
    # 1. Solicitar el enlace de descarga
    response = requests.get(url_base, headers=headers)
    
    if response.status_code == 200:
        meta_datos = response.json()
        print(meta_datos) 
        print(f"Respuesta AEMET: {meta_datos.get('descripcion')}")
       
        if meta_datos.get('estado') == 200:
            url_final = meta_datos.get('datos')
            
            # 2. Descargar el JSON con los datos de las estaciones
            datos_response = requests.get(url_final)
            return datos_response.json()
        else:
            print(f"Error AEMET: {meta_datos.get('descripcion')}")
    else:
        print(f"Error de red: {response.status_code}")
    return None

def cargar_datos_climaticos_a_db(climaticos):
    # Aquí iría la lógica para conectar a la base de datos y cargar los datos climáticos
    data = []
    for dia in climaticos:
        data.append({"fecha"       : dia.get("fecha"),
                     "indicativo"  : dia.get("indicativo"),
                     "nombre"      : dia.get("nombre"),
                     "provincia"   : dia.get("provincia"),
                     "altitud"     : dia.get("altitud"),
                     "tmed"        : dia.get("tmed"),
                     "prec"        : dia.get("prec", "0,00"), 
                     "tmin"        : dia.get("tmin"),
                     "horatmin"    : dia.get("horatmin"),
                     "tmax"        : dia.get("tmax"),
                     "horatmax"    : dia.get("horatmax"),
                     "dir"         : dia.get("dir"),
                     "velmedia"    : dia.get("velmedia"),
                     "racha"       : dia.get("racha"),
                     "horaracha"   : dia.get("horaracha"),
                     "sol"         : dia.get("sol"),
                     "presMax"     : dia.get("presMax"),
                     "horaPresMax" : dia.get("horaPresMax"),
                     "presMin"     : dia.get("presMin"),
                     "horaPresMin" : dia.get("horaPresMin"),
                     "hrMedia"     : dia.get("hrMedia"),
                     "hrMax"       : dia.get("hrMax"),
                     "horaHrMax"   : dia.get("horaHrMax"),
                     "hrMin"       : dia.get("hrMin"),
                     "horaHrMin"   : dia.get("horaHrMin")
                     })
    
    df = pd.DataFrame(data)
    
    df["fecha"]       = pd.to_datetime(df["fecha"])
    df["indicativo"]  = df["indicativo"].astype(str)
    df["nombre"]      = df["nombre"].astype(str)
    df["provincia"]   = df["provincia"].astype(str)
    df["altitud"]     = pd.to_numeric(df["altitud"])
    df["tmed"]        = df["tmed"].str.replace(",", ".").astype(np.float32)
    df["prec"]        = df["prec"].fillna("0.00").str.replace(",", ".").replace("", "0.00").astype(np.float32)
    df["tmin"]        = df["tmin"].str.replace(",", ".").astype(np.float32)
    df["horatmin"]    = pd.to_datetime(df["horatmin"], format="%H:%M", errors="coerce").dt.time
    df["tmax"]        = df["tmax"].str.replace(",", ".").astype(np.float32)
    df["horatmax"]    = pd.to_datetime(df["horatmax"], format="%H:%M", errors="coerce").dt.time
    df["dir"]         = df["dir"].astype(str)
    df["velmedia"]    = df["velmedia"].str.replace(",", ".").astype(np.float32)
    df["racha"]       = df["racha"].str.replace(",", ".").astype(np.float32)
    df["horaracha"]   = df["horaracha"].astype(str)
    df["sol"]         = df["sol"].str.replace(",", ".").astype(np.float32)
    df["presMax"]     = df["presMax"].str.replace(",", ".").astype(np.float32)
    df["horaPresMax"] = pd.to_datetime(df["horaPresMax"], format="%H", errors="coerce").dt.time
    df["presMin"]     = df["presMin"].str.replace(",", ".").astype(np.float32)
    df["horaPresMin"] = pd.to_datetime(df["horaPresMin"], format="%H", errors="coerce").dt.time
    df["hrMedia"]     = df["hrMedia"].str.replace(",", ".").astype(np.float32)
    df["hrMax"]       = df["hrMax"].str.replace(",", ".").astype(np.float32)
    df["horaHrMax"]   = pd.to_datetime(df["horaHrMax"], format="%H:%M", errors="coerce").dt.time
    df["hrMin"]       = df["hrMin"].str.replace(",", ".").astype(np.float32)
    df["horaHrMin"]   = pd.to_datetime(df["horaHrMin"], format="%H:%M", errors="coerce").dt.time
    
   
    return df



# # Datos Configuración para la consulta a la API de AEMET
idema = "B228"                       # Ejemplo: Código de la estación de Palma de Mallorca
fecha_ini = "2026-05-01T00:00:00UTC"  # Fecha de inicio
fecha_fin = "2026-05-15T00:00:00UTC"  # Fecha de fin

# # Ejecución y visualización
climaticos = obtener_valores_climaticos(fecha_ini, fecha_fin, idema)

# Ejemplo: total de estaciones
print(f"Total dias de la estacion: {len(climaticos)}")   

# Ejemplo: del JSON completo de los dias de la estacion
print(climaticos)



# if climaticos:
#     # Ejemplo: Imprimir los dias de una estacion con sus datos climáticos
#     for dia in climaticos:

#         #Ejemplo de campos de JSON por dia estacion 
#         #{'fecha': '2026-05-01', 
#         # 'indicativo': 'B228', 
#         # 'nombre': 'PALMA, PUERTO', 
#         # 'provincia': 'BALEARES', 
#         # 'altitud': '3', 
#         # 'tmed': '19,6', 
#         # 'prec': '0,0', 
#         # 'tmin': '17,5', 
#         # 'horatmin': '05:30', 
#         # 'tmax': '21,8', 
#         # 'horatmax': '10:10', 
#         # 'dir': '05', 
#         # 'velmedia': '1,7', 
#         # 'racha': '6,4', 
#         # 'horaracha': 'Varias', 
#         # 'sol': '0,8', 
#         # 'presMax': '1023,0', 
#         # 'horaPresMax': '23', 
#         # 'presMin': '1016,8', 
#         # 'horaPresMin': '03', 
#         # 'hrMedia': '83', 
#         # 'hrMax': '92', 
#         # 'horaHrMax': '05:20', 
#         # 'hrMin': '72', 
#         # 'horaHrMin': 'Varias'}, 



#         print(f"fecha: {dia['fecha']} | indicativo: {dia['indicativo']} | nombre: {dia['nombre']} | provincia: {dia['provincia']} | temperatura media: {dia['tmed']} | temperatura mínima: {dia['tmin']} | temperatura máxima: {dia['tmax']} | velocidad media del viento: {dia['velmedia']} | racha máxima de viento: {dia['racha']} | horas de sol: {dia['sol']} | presión máxima: {dia['presMax']} | presión mínima: {dia['presMin']} | humedad relativa media: {dia['hrMedia']}")
```
